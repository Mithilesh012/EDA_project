# What are the most important features affecting house prices?

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[['Price','Bedrooms', 'Bathrooms','Price']], bins=10, color='skyblue', edgecolor='black')

plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30 )#,kde=True
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# How do house prices vary by neighborhood or location?

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='Neighborhood', y='Price')
plt.title('Price Distribution by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

# Are there any outliers in the data, and what could be the reason for them?

```python
plt.figure(figsize=(12, 6))

sns.boxplot(x=df['Price'])
plt.title('Box Plot of House Prices')
plt.xlabel('Price')
plt.show()
```

# What is the relationship between house size (square footage) and price?

```python
sns.scatterplot(data=df, x='SquareFeet', y='Price', color='blue', edgecolor='black')
plt.title('House Size vs. Price')
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
neighborhood_sqft_sum = df.groupby('Neighborhood')['SquareFeet'].sum()
print(neighborhood_sqft_sum)
```

```python
neighborhood_sqft_sum = df.groupby('Neighborhood')['SquareFeet'].sum().reset_index()
print(neighborhood_sqft_sum)
plt.figure(figsize=(12, 8))
sns.barplot(data=neighborhood_sqft_sum, x='Neighborhood', y='SquareFeet', palette='viridis')
plt.xticks(rotation=90)
plt.title('Total Square Footage by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Total Square Footage')

plt.show()
```
